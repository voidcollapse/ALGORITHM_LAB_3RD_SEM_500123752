```python
import matplotlib.pyplot as plt
```

```python
# Data for Array Type, Recursive Time and Iterative Time
array_type = ['Sorted', 'Reverse Sorted', 'Random', 'Uniform', 'Balanced']
recursive_time = [0.241000, 0.224000, 0.150000, 0.000000, 0.125000]
iterative_time = [0.083000, 0.068000, 0.000000, 0.000000, 0.042000]
```

```python
# Create a line plot for both Recursive and Iterative Insertion Times
plt.figure(figsize=(10, 6))
plt.plot(array_type, recursive_time, label="Recursive Method", marker='o', linestyle='-', color='b')
plt.plot(array_type, iterative_time, label="Iterative Method", marker='o', linestyle='-', color='r')

# Add labels and title
plt.xlabel("Array Type")
plt.ylabel("Time (seconds)")
plt.title("Comparison of Recursive and Iterative Insertion Method")

# Add legend
plt.legend()

# Add grid for better readability
plt.grid(True)

# Show the plot
plt.show()
```
