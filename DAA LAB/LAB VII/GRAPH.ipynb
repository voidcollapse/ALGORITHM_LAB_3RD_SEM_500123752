```python
import matplotlib.pyplot as plt
```

```python
# Test case data
test_cases = [1, 2, 3, 4, 5]

# Time data for Greedy and DP approaches (in milliseconds)
greedy_times = [0.000000, 0.000000, 0.000000, 0.000103, 0.000069]  # Greedy execution time in ms
dp_times = [0.000271, 0.000205, 0.000207, 0.000206, 0.000204]  # DP execution time in ms

# Value data for Greedy and DP approaches
greedy_values = [45, 50, 50, 25, 45]  # Greedy knapsack value
dp_values = [50, 50, 50, 50, 50]  # DP knapsack value
```

```python
# Time Comparison - Line Plot
plt.figure(figsize=(12, 6))

# Line plot for execution time comparison
plt.subplot(1, 2, 1)
plt.plot(test_cases, greedy_times, label="Greedy Time", marker='o', linestyle='-', color='b')
plt.plot(test_cases, dp_times, label="DP Time", marker='o', linestyle='-', color='r')
plt.xlabel('Test Case')
plt.ylabel('Execution Time (ms)')
plt.title('Time Comparison')
plt.legend()

# Value Comparison - Bar Plot
plt.subplot(1, 2, 2)
width = 0.35  # Width of bars
x = [tc - width/2 for tc in test_cases]  # Adjust x positions for Greedy bars
plt.bar(x, greedy_values, width=width, label='Greedy Value', color='b')

x = [tc + width/2 for tc in test_cases]  # Adjust x positions for DP bars
plt.bar(x, dp_values, width=width, label='DP Value', color='r')

plt.xlabel('Test Case')
plt.ylabel('Knapsack Value')
plt.title('Value Comparison')
plt.legend()

# Show the plots
plt.tight_layout()
plt.show()
```
