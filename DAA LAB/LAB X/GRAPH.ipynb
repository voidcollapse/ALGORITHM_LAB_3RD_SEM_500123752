```python
import matplotlib.pyplot as plt
```

```python
# Test case data
test_cases = [1, 2, 3, 4, 5]
naive_times = [0.0, 1.0, 222.0, 110.0, 0.0]
rabin_karp_times = [1.0, 2.0, 631.0, 327.0, 0.0]
kmp_times = [0.0, 1.0, 322.0, 150.0, 0.0]
```

```python
# Plotting the execution times
plt.figure(figsize=(10, 6))
plt.plot(test_cases, naive_times, label="Naive", marker='o', linestyle='-', color='blue')
plt.plot(test_cases, rabin_karp_times, label="Rabin-Karp", marker='o', linestyle='-', color='red')
plt.plot(test_cases, kmp_times, label="KMP", marker='o', linestyle='-', color='green')

# Adding titles and labels
plt.title("Execution Time Comparison for String Matching Algorithms")
plt.xlabel("Test Cases")
plt.ylabel("Execution Time (ms)")
plt.xticks(test_cases)
plt.legend()

# Show the plot
plt.grid(True)
plt.show()
```
