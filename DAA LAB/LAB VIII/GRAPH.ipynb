```python
import matplotlib.pyplot as plt
```

```python
# Set sizes and corresponding execution times from the test cases
set_sizes = [5, 10, 15, 20, 25]  # Sizes from the test cases
execution_times = [0.000000, 0.001000, 0.016000, 0.136000, 0.048000]  # Execution times in seconds for each case
```

```python
# Plotting the line graph
plt.figure(figsize=(8, 6))
plt.plot(set_sizes, execution_times, marker='o', linestyle='-', color='r', label='Execution Time')

# Adding titles and labels
plt.title('Performance of Sum of Subsets Algorithm', fontsize=14)
plt.xlabel('Set Size', fontsize=12)
plt.ylabel('Execution Time (seconds)', fontsize=12)
plt.legend()

# Show the plot
plt.show()
```
