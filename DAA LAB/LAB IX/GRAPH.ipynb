```python
import matplotlib.pyplot as plt
```

```python
# Test case data
test_cases = [1, 2, 3, 4, 5]

# Time data for DP, Backtracking, and Branch & Bound approaches (in milliseconds)
dp_times = [0.000000, 0.000048, 0.000000, 0.000100, 0.000204]  # DP execution time in ms
backtracking_times = [0.000000, 0.000000, 0.000802, 0.000087, 0.000121]  # Backtracking execution time in ms
branch_bound_times = [0.000646, 0.000000, 0.000200, 0.000108, 0.000200]  # Branch & Bound execution time in ms

# Value data for DP, Backtracking, and Branch & Bound approaches
dp_values = [50, 50, 50, 50, 50]  # DP knapsack value
backtracking_values = [50, 50, 50, 50, 50]  # Backtracking knapsack value
branch_bound_values = [50, 50, 50, 50, 50]  # Branch & Bound knapsack value
```

```python
# Time Comparison - Line Plot
plt.figure(figsize=(12, 6))

# Line plot for execution time comparison
plt.subplot(1, 2, 1)
plt.plot(test_cases, dp_times, label="DP Time", marker='o', linestyle='-', color='r')
plt.plot(test_cases, backtracking_times, label="Backtracking Time", marker='o', linestyle='-', color='b')
plt.plot(test_cases, branch_bound_times, label="Branch & Bound Time", marker='o', linestyle='-', color='g')
plt.xlabel('Test Case')
plt.ylabel('Execution Time (ms)')
plt.title('Time Comparison')
plt.legend()

# Value Comparison - Bar Plot
plt.subplot(1, 2, 2)
width = 0.25  # Width of bars
x = [tc - width for tc in test_cases]  # Adjust x positions for DP bars
plt.bar(x, dp_values, width=width, label='DP Value', color='r')

x = [tc for tc in test_cases]  # Adjust x positions for Backtracking bars
plt.bar(x, backtracking_values, width=width, label='Backtracking Value', color='b')

x = [tc + width for tc in test_cases]  # Adjust x positions for Branch & Bound bars
plt.bar(x, branch_bound_values, width=width, label='Branch & Bound Value', color='g')

plt.xlabel('Test Case')
plt.ylabel('Knapsack Value')
plt.title('Value Comparison')
plt.legend()

# Show the plots
plt.tight_layout()
plt.show()
```
