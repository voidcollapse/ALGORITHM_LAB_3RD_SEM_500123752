```python
import matplotlib.pyplot as plt
```

```python
# Data for matrix sizes, traditional times, and Strassen times
sizes = ['64x64', '128x128', '256x256', '512x512', '1024x1024', '2048x2048']
traditional_times = [0.007000, 0.013000, 0.146000, 0.583000, 6.785000, 87.368000]
strassen_times = [0.000000, 0.012000, 0.077000, 0.409000, 2.938000, 19.229000]
```

```python
# Create a line plot for both Traditional and Strassen times
plt.figure(figsize=(10, 6))
plt.plot(sizes, traditional_times, label="Traditional Method", marker='o', linestyle='-', color='b')
plt.plot(sizes, strassen_times, label="Strassen Method", marker='o', linestyle='-', color='r')

# Add labels and title
plt.xlabel("Matrix Size")
plt.ylabel("Time (seconds)")
plt.title("Comparison of Traditional and Strassen Matrix Multiplication")

# Add legend
plt.legend()

# Add grid for better readability
plt.grid(True)

# Show the plot
plt.show()
```
